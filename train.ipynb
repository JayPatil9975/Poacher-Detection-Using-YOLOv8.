```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(data="dataset.yaml", epochs=10, imgsz=640, batch=8, device="cpu")
```

```python
metrics = model.val()  
print(metrics)
```

```python
results = model("10-yr-wanted-poaching-mastermind-ZAWA-LATF.jpg")
results[0].show()
boxes = results[0].boxes.xyxy[0]
print(boxes)
```
